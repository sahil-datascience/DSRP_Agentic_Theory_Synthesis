# Vectorise the Pilot Set


```python
from langchain_docling import DoclingLoader
from docling.chunking import HybridChunker
from pathlib import Path

# Create vector storage folder for pilot study
vector_storage_path = Path(r"C:\Users\sahil\OneDrive\PhD\3. Empirical Study\Methodological Workflow\Pilot_Evaluation\pilot_study_10_vectors")
vector_storage_path.mkdir(parents=True, exist_ok=True)

print(f"Vector storage location: {vector_storage_path.absolute()}")
print(f"Number of PDFs to vectorize: {len(random_pdfs)}")
```

```python
from langchain_core.documents import Document
import os
import uuid


def convert_docling_chunks(doc_splits):

    clean_documents = []

    for i, doc in enumerate(doc_splits):

        dl_meta = doc.metadata.get("dl_meta", {})
        origin = dl_meta.get("origin", {})
        headings = dl_meta.get("headings", [])
        doc_items = dl_meta.get("doc_items", [])

        # Extract page number safely
        page_no = None
        if doc_items:
            prov = doc_items[0].get("prov", [])
            if prov:
                page_no = prov[0].get("page_no")

        filename = origin.get("filename")
        paper_id = os.path.splitext(filename)[0]

        clean_metadata = {
            "paper_id": paper_id,
            "filename": filename,
            "section_heading": headings[0] if headings else None,
            "page_no": page_no,
            "modality": "text",  # later you can detect table/figure
            "chunk_id": f"{paper_id}_{i}"
        }

        clean_doc = Document(
            page_content=doc.page_content,
            metadata=clean_metadata
        )

        clean_documents.append(clean_doc)

    return clean_documents
```

```python
# Process all PDFs in random_pdfs
all_documents = []

for pdf_path in random_pdfs:
    try:
        print(f"Processing: {pdf_path.name}...")
        loader = DoclingLoader(file_path=str(pdf_path), chunker=HybridChunker())
        docs = loader.load()
        documents = convert_docling_chunks(docs)
        all_documents.extend(documents)
        print(f"  ✓ Loaded {len(documents)} chunks from {pdf_path.name}")
    except Exception as e:
        print(f"  ✗ Error processing {pdf_path.name}: {e}")

print(f"\n📊 Total chunks extracted: {len(all_documents)}")
```

```python
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings

def store_vectorized_documents(all_documents, vector_storage_path):
    """
    Store all documents in a unified Chroma vector store for the pilot study.
    Each document maintains its paper_id in metadata for easy filtering per-paper in RAG workflows.
    """
    
    embedding_function = OpenAIEmbeddings(model="text-embedding-3-small")
    
    # Create unified vector store for all pilot study papers
    collection_name = "pilot_study_10"
    
    vectorstore = Chroma(
        collection_name=collection_name,
        embedding_function=embedding_function,
        persist_directory=str(vector_storage_path)
    )
    
    # Add all documents (metadata includes paper_id for filtering)
    vectorstore.add_documents(all_documents)
    vectorstore.persist()
    
    print(f"✅ Vectorized {len(all_documents)} chunks")
    print(f"📁 Vector store location: {vector_storage_path.absolute()}")
    print(f"📚 Collection name: {collection_name}")
    print(f"💾 Papers stored: {len(set(doc.metadata['paper_id'] for doc in all_documents))}")
    
    return vectorstore
```

```python
# Store all vectorized documents
vectorstore = store_vectorized_documents(all_documents, vector_storage_path)
```

```python
# Helper function to retrieve documents for a specific paper during RAG workflow
def retrieve_paper_documents(vectorstore, paper_id, query, k=5):
    """
    Retrieve documents for a specific paper using metadata filtering.
    Useful for per-paper agentic RAG workflows.
    
    Args:
        vectorstore: Chroma vectorstore
        paper_id: Paper identifier (filename without extension)
        query: Search query
        k: Number of results
    
    Returns:
        List of filtered documents
    """
    from langchain_core.documents import Document
    
    # Perform similarity search
    results = vectorstore.similarity_search(query, k=k*2)
    
    # Filter by paper_id
    paper_docs = [doc for doc in results if doc.metadata.get('paper_id') == paper_id]
    
    return paper_docs[:k]

print("✅ Helper function ready for per-paper retrieval in agentic RAG workflows")
```
